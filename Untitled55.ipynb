```python
#For Question 3
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

%matplotlib inline
sb.set_style("whitegrid")
```

```python
dogs_df = pd.read_csv("dogs.csv")
```

```python
dogs_df.head()
```

```python
#Lets check for the missing values
dogs_df.isna().sum()
```

```python
# From the above it's clear that we are having a lot of missing values
# However, we need to do a careful analysis of the columns before deciding to discard any of the data. 

# breed_secondary column - Since 65% values the values are missing and breed_primary is already having all the breeds, 
# dropping breed_secondary should not cause any big problems.

# For colors, we can merge primary, secondary and tertiary into one column by substituting secondary/tertiary color
# wherever there is missing value in primary.   
```

```python
dogs_df = dogs_df.drop(['breed_secondary','tags','contact_zip','color_tertiary','name', 'stateQ'], axis=1)
```

```python
dogs_df.isna().sum()
```

```python
#For declawed, env_children, env_dogs, env_cats , we can replace the null values with false

dogs_df['declawed'].fillna(False, inplace=True)
dogs_df['env_children'].fillna(False, inplace=True)
dogs_df['env_cats'].fillna(False, inplace=True)
dogs_df['env_dogs'].fillna(False, inplace=True)
```

```python
dogs_df['coat'].fillna(method = 'ffill', inplace=True)
```

```python
dogs_df.color_primary.fillna(dogs_df.color_secondary, inplace=True)
del dogs_df['color_secondary']
```

```python
dogs_df['color_primary'].fillna("Unknown", inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
dogs_df.groupby('age').count()['breed_primary'].plot(ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
dogs_df.groupby('age')['breed_primary'].value_counts().plot()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
dogs_df.groupby('contact_city')['breed_primary'].value_counts().plot()
```
